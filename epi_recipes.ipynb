# Data Cleaning: Removing Null Values and Duplicates

```python
# To check whether title column has any blank/null values or not
df.title.isna().sum()
#df.calories.isna().sum()
```

```python
df.shape
```

```python
#To drop dupliactes from primary column 'title'
df.drop_duplicates(subset=['title'],inplace = True)
```

```python
df.shape
```

```python
# To check whether calories column have any blank/null values or not
df.calories.isnull().sum()
```

```python
# removing the rows where calories value is more than 10000 and also retaining blank rows for calories
df = df[(df['calories']<10000) | (df['calories'].isnull() ==1)]
```

```python
# Calculating the average/mean value for calories column
calories_mean = df.calories.mean()
calories_mean
```

```python
# replace blank values in calories column with its mean value
#df.calories.fillna(calories_mean, inplace=True)
df['calories']=df['calories'].fillna(calories_mean)
df.calories.isnull().sum()
df.calories.describe()
```

```python
# To check whether protein column have any blank/null values or not
df.protein.isnull().sum()
```

```python
# Calculating the average/mean value for protein column
protein_mean=df.protein.mean()
protein_mean
```

```python
# # replace blank values in protein column with its mean value
df['protein']=df['protein'].fillna(protein_mean)
df.protein.isnull().sum()
```

```python
# To check if fat column have blank/null values or not
df.fat.isnull().sum()
```

```python
# calculating the mean for fat column
fat_mean=df.fat.mean()
fat_mean
```

```python
# replace blank values in fat column with its mean value
df['fat']=df['fat'].fillna(fat_mean)
df.fat.isnull().sum()
```

```python
# To check whether sodium column have any blank/null values
df.sodium.isnull().sum()
```

```python
# Calculating the mean value for sodium column
sodium_mean=df.sodium.mean()
sodium_mean
df.sodium.describe()
```

```python
# replace blank values in sodium column with its mean value
df['sodium']=df['sodium'].fillna(sodium_mean)
df.sodium.isnull().sum()
```

```python
# Removing columns from dataset which appear in less than 10 recipies 
c=0;
for col in df.iloc[:,6:].columns:
    if len(df[df[col]==1])<10:
      df=df.drop(col,axis=1)
      print(col,end=" ")
      c+=1

print(" no of columns dropped:",c)
    
```

```python
# To show correlation between rating,calories,protein,fat and sodium

correlation_mat=df.iloc[:,1:6].corr().abs()
correlation_mat
```

```python
# To list out the common ingerdients in high rated(5) recipes

highrate_recipes = df[df['rating'] == 5.0]
highrate_recipes_ingerdients = list()
i=0
for col in highrate_recipes.iloc[:,6:].columns:
    if len(highrate_recipes[highrate_recipes[col]==1])>250:
        #print(col)
        highrate_recipes_ingerdients.append(col)
        i+=1
print(f"There are total {i} no of common ingredients in High-rated recipes:")        
print("The list of common ingredients in High-Rated Recipes is as follows:")       
for col in highrate_recipes_ingerdients:        
    print(col,end=" ")   
```

```python
# To show the correlation between preparation time and rating

print("Correlation between preparation time and rating:")
print(df['rating'].corr(df['advance prep required']))
```

```python
# To show the top 10 most common tags

ingredient_counts=df.iloc[:,9:].sum()
ingredient_count_sorted=ingredient_counts.sort_values(ascending=False)
top_10_ingredients=ingredient_count_sorted[:10]
plt.figure(figsize=(10,6))
top_10_ingredients.plot(kind="bar")
plt.title("Top 10 most common tags")
plt.xlabel("Tags")
plt.ylabel("no of recieps")
plt.show()
```
